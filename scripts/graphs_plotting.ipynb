```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
def get_dataset_info(dataset_name):
    os.chdir(dataset_name)
    csv_files = [file for file in os.listdir() if file.endswith('.csv')]
    dataframes = []
    approaches = []
    for file in csv_files:
        approaches.append(file[:-4])
        df = pd.read_csv(file)
        dataframes.append(df)
    os.chdir('..')
    return [dataframes, approaches]
```

```python
dataframes, approaches = get_dataset_info('SiouxFalls')
```

```python
def plot_datasets_graphs(dataframes, approaches):
    for [df, name] in zip(dataframes, approaches):
        plt.plot(df["Time"], df["RGAP"], label = name)
    plt.legend()
    plt.yscale('log')
    plt.grid(axis='x', which='major')
    plt.grid(axis='y', which='major')
    plt.yticks([10 ** i for i in range(-18, 0, 3)])
    plt.xlabel('sec')
    plt.ylabel('RGAP')
    plt.show()
```

```python
plot_datasets_graphs(*get_dataset_info('SiouxFalls'))
```

```python
dataframes, approaches = get_dataset_info('SiouxFalls')
plot_datasets_graphs(*get_dataset_info('SiouxFalls'))
```
